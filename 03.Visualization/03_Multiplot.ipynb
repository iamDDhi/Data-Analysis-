## 여러개의 그림 그리기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = 'Malgun Gothic'
```

- Figure 객체

```python
np.random.seed(2021)
plt.figure(figsize=(10,2))
plt.title('figure size:(10,2)')
plt.plot(np.random.randn(100))
plt.grid()
plt.show()
```

- Axes 객체와 subplot

```python
x1 = np.linspace(0.0, 5.)
x2 = np.linspace(0., 2.)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
plt.title('2 sublot')
ax1 = plt.subplot(2, 1, 1)
ax1.plot(x1, y1, 'yo-')
ax1.set_ylabel('Damped oscilliation')

ax2 = plt.subplot(2, 1, 2)
ax2.plot(x2, y2, 'r.-')
ax2.set_ylabel('Undamped')


plt.xlabel('time (sec)')
plt.show()
```

```python
ax1 = plt.subplot(211)
plt.title('2 Subplots')
ax1.plot(x1, y1, 'yo-')
ax1.set_ylabel('Damped oscilliation')

ax2 = plt.subplot(212)
ax2.plot(x2, y2, 'r.-')
ax2.set_ylabel('Undamped')


plt.xlabel('time (sec)')
ax1.grid(); ax2.grid()
plt.show()
```

```python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
np.random.seed(0)

fig, axes = plt.subplots(2,2)
for i in range(4):
    ax = axes[i//2, i%2]    #행인덱스-i//2 / 열 인덱스-i%2
    ax.plot(np.random.rand(5))
    ax.set_title(f'axes {i+1}')
    ax.grid()

plt.tight_layout()
plt.show()
```
